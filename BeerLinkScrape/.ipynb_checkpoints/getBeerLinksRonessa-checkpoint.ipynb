```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
# read in beerstyles csv
df = pd.read_csv('styleLinks.csv')
```

```python
df.head()
```

```python
csv_names = list(df['csv_name'])
match_names = list(df['match_name'])
style_urls = list(df['style_link'])
style_names = list(df['style_name'])
```

```python


# url = "https://www.beeradvocate.com/beer/style/38/"
```

```python
loop_count = 1
count = 0

beer_list = []
ranking_list = []
beer_links = []

all_beers_links = []

for x in range(len(style_urls)):
    count = 0
    flag = True
    
    beer_list = []
    ratings_list = []
    beer_links = []
    
    while flag:

        suffix = f"?sort=revsD&start={count}"
        style_url = style_urls[x] + suffix

        html = requests.get(style_url)
        soup = BeautifulSoup(html.text, 'html.parser')


        results = soup.find_all("td", class_="hr_bottom_light")
#         print(results)
        print(style_url, x, count, len(results))

        if len(results) == 0:
            print("style: "+ style_names[x] + "finished")
            flag = False
            break

        
        current_ratings = [int(x.text.replace(",", "")) for x in results[3::5] if int(x.text.replace(",", "")) > 10]
        print(len(current_ratings))
        if len(current_ratings) > 0:
            beer_list = [x.a.text for x in results[0::5]]
            ratings_list = current_ratings
            beer_links = [x.a['href'] for x in results [0::5]]
            scores = [x.text for x in results[4::5]]
            abv = [x.text for x in results[2::5]]
            brewery = [x.text for x in results[1::5]]
        
            for i in range(len(ratings_list)):
                beer_dict = {
                    "beer_name": beer_list[i],
                    "beer_style": style_names[x],
                    "beer_url": "https://www.beeradvocate.com" + beer_links[i],
                    "csv_name": csv_names[x],
                    "match_name": match_names[x],
                    "style_url": style_urls[x],
                    "num_ratings": ratings_list[i],
                    'score': scores[i],
                    'abv': abv[i],
                    'brewery': brewery[i]
                }
                all_beers_links.append(beer_dict)
#                 print(loop_count)
#                 loop_count += 1
        else:
            print("style: "+ style_names[x] + "finished bc reach num ratings of 10")
            flag = False
            break
        
        
        if (len(beer_list) == len(ratings_list)) and (len(ratings_list)  == len(beer_links)):
            print(f"All good with {style_names[x]}: {count} to {count+50}")
        else:
            print(f"Something Wrong with {style_names[x]}: {count} to {count+50}")
        
       
        
        count += 50
```

```python
allbeerlinks = pd.DataFrame(all_beers_links)
```

```python
allbeerlinks.iloc[-1]
```

```python
allbeerlinks['num_ratings'] = pd.to_numeric(allbeerlinks['num_ratings'], errors = 'coerce')
allbeerlinks_filtered = allbeerlinks[(allbeerlinks["num_ratings"] >= 10)]
```
