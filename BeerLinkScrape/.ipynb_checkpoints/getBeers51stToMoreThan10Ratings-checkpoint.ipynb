```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
# read in beerstyles csv
df = pd.read_csv('styleLinks.csv')
```

```python
df.head()
```

```python
baseUrl = 'https://www.beeradvocate.com'
beerlinks = []
beernames = []
beerstyles = []


# urls = ['https://www.beeradvocate.com/beer/style/38/','https://www.beeradvocate.com/beer/style/21/']
# for url in urls:
for index, row in df.iterrows():
#     print(row[2], row[3])
    url = row[2]
    style_name = row[3]
    
    # start with the second page of the beer style list (beers 51-100) and get the beers and urls (by count of ratings)
    flag = True
    beerSet = 50 # paginated by 50s
    
    while flag:
        
        suffix = '?sort=revsD&start=' + str(beerSet)

        # we'll loop through the pages for each style here
        try:
            response = requests.get(url + suffix)
        except:
            print(url, 'this page does not exist')

        # Create BeautifulSoup object; parse with 'html.parser'
        soup = BeautifulSoup(response.text, 'html.parser')

        beerTable = soup.find('table')

        tds = soup.find_all('td', {'align':'left'})
    
        try:
            
            hrefs = soup.select('tr td a[href]')

            if beerSet == 50:
                # only get the last page the first time through
                lastPage = hrefs[6:7]
                for last in lastPage:
                    print(last['href'])
                    lastUrl = last['href']
                    pos = lastUrl.find('start=')
                    maxPage = int(lastUrl[pos+6:])
                    print("maxPage",maxPage)

    #                 flag = False
    #                 check the rating count for the first beer on the list
                ratingCount = tds[5].text
                print("rating", beerSet, ratingCount)
            elif beerSet >= 100 and beerSet < maxPage:
                # the navigation links change by page so need to adjust where we are getting this from
                ratingCount = tds[5].text
                print("rating", beerSet, ratingCount)

            if int(ratingCount) >= 10:

                #get the count of how many beers are on the page
                pageBeerCount = len(hrefs) # of navigational links per page changes based on which page you are on
                print("pageBeerCount", pageBeerCount)

                if beerSet == 50:
                    beerHrefs = hrefs[12:(pageBeerCount-7)]
                    print(50,beerHrefs)
                elif beerSet == (maxPage - 50):
                    beerHrefs = hrefs[12:(pageBeerCount-7)]
                elif beerSet == maxPage:
                    beerHrefs = hrefs[4:(pageBeerCount-4)]
                else:
                    beerHrefs = hrefs[13:(pageBeerCount-8)]

                # then loop through and get every other (first is the beer, second is the brewery)
                count = 0
                for beerHref in beerHrefs:
                    if (count % 2 == 0):
                        beerlink = beerHref['href']
                        beername = beerHref.text
            #             print(beerlink, beername)
                        beerlinks.append(baseUrl + beerlink)
                        beernames.append(beername)
                        beerstyles.append(style_name)
                    count += 1

                beerSet += 50
            else:
                print("rating count is less than 10 - done with this style", url)
                flag = False

        except:
            print(url, "error", tds[6].text)
            flag = False
            
```

```python
# create a df of all the beers to collect the rest of their features
beerdf = pd.DataFrame({"style":beerstyles,"names":beernames,"url":beerlinks})
beerdf.head()
```
