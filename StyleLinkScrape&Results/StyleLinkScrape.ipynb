```python
import csv
from bs4 import BeautifulSoup
import pandas as pd
import requests
import re
```

```python
# file with beer features
file = "beerFeatures.csv"

# list for beer types to pull
beer_styles = []

#read csv and append beer style to list
with open(file) as csv_file:
    csv_read = csv.reader(csv_file)
    
    next(csv_read)
    
    for row in csv_read:
        beer_style = re.sub(r'\W+', '', row[0].lower())
        beer_styles.append(beer_style)
print(beer_styles)
print(len(beer_styles))
```

```python
# beer advocate styles url
styles_url = "https://www.beeradvocate.com/beer/style/"

# response from site
response = requests.get(styles_url)

#create Beautiful soup object and parse
soup = BeautifulSoup(response.text, 'html.parser')
results = soup.find_all('table')

# find all 'tds' in results, only index 0 holds correct table
tds = results[0].find_all('td')

# only index 0 and 4 hold the two necessary columns of beer styles
target_indexes = [0, 4]
```

```python
# list for all beer styles from beer advocate
all_styles = []
base_url = "https://www.beeradvocate.com"

# get a dict to hold style and link
for x in target_indexes:
    ind_beers = tds[x].find_all('a')
    for beer in ind_beers:
        beer_text = re.sub(r'\W+', '', beer.text.lower())
        beer_dict = {"style_link": base_url + beer['href'], "style_name": beer.text, "match_name": beer_text}
        all_styles.append(beer_dict)
```

```python
# # beer styles pulled
len(all_styles)
```

```python
# list for matching styles
matching_styles = []

# match each style in our list as long as the style in part of the style from beer advocate
for style in beer_styles:
    for x in all_styles:
        if style in x['match_name']:
            x['csv_name'] = style
            matching_styles.append(x)
        
len(matching_styles)
```

```python
beer_styles_df = pd.DataFrame(matching_styles).drop_duplicates()
len(beer_styles_df)
```

```python
beer_styles_df = beer_styles_df[beer_styles_df['style_name'] != "Eisbock"]
```
