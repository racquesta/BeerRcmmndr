```python
import pandas as pd
import re
```

```python
beer_features = pd.read_csv('beerFeatures.csv')
style_links = pd.read_csv('styleLinks.csv')
```

```python
beer_features.head()
```

```python
del beer_features['Examples']
```

```python
del beer_features['Pairing']
```

```python
del beer_features['Store Temp']
```

```python
beer_features['csv_name'] = ""

for index, row in beer_features.iterrows():
    beer_features.set_value(index, 'csv_name', re.sub(r'\W+', '', row['Style'].lower()))

```

```python
style_links = style_links[['csv_name', 'style_name', 'match_name']]
```

```python
len(beer_features)
```

```python
len(style_links)
```

```python
beer_features = beer_features.merge(style_links, on = "csv_name", how = "inner")
len(beer_features)
```

```python
del beer_features['Style']
```

```python
del beer_features['Description']
```

```python
del beer_features['Descriptive Words']
```
